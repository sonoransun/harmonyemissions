# 03 — Laser-wakefield betatron X-ray source

An electron oscillating inside the ion-bubble of an LWFA emits synchrotron-like radiation with critical frequency ω_c ≈ (3/2) γ³ ω_β² r_β / c.

```python
import matplotlib.pyplot as plt
from harmonyemissions import Laser, Target, simulate
from harmonyemissions.viz import plot_spectrum
```

```python
laser = Laser(a0=2.0, wavelength_um=0.8, duration_fs=30.0, envelope='sin2')
target = Target.underdense(n_over_nc=0.001, bubble_radius_um=10.0, electron_energy_mev=500.0, betatron_amplitude_um=1.0)
result = simulate(laser, target, model='betatron')
print({k: v for k, v in result.diagnostics.items()})
```

```python
fig, ax = plt.subplots(figsize=(7, 4))
plot_spectrum(result, ax=ax, show_fit=False)
ax.axvline(result.diagnostics['omega_c_over_omega0'], color='red', lw=0.8, ls='--', label='ω_c')
ax.set_title(f'LWFA betatron, γ_e={result.diagnostics["gamma_e"]:.0f}')
ax.legend()
plt.show()
```

```python
for E in [100, 200, 500, 1000]:
    r = simulate(laser, Target.underdense(0.001, electron_energy_mev=E), model='betatron')
    print(f'E={E} MeV  ω_c/ω₀={r.diagnostics["omega_c_over_omega0"]:.1f}  photon≈{r.diagnostics["photon_energy_keV_at_cutoff"]:.2f} keV')
```
