# 01 — Surface HHG via the Relativistic Oscillating Mirror

This notebook walks through a single ROM run and shows the harmonic spectrum, plateau fit, and Doppler-compressed reflected field.

```python
import matplotlib.pyplot as plt

from harmonyemissions import Laser, Target, simulate
from harmonyemissions.config import NumericsConfig
from harmonyemissions.viz import plot_pulse, plot_spectrum
```

```python
laser = Laser(a0=15.0, wavelength_um=0.8, duration_fs=8.0, cep=0.0, envelope='gaussian')
target = Target.overdense(n_over_nc=200.0, gradient_L_over_lambda=0.05)
numerics = NumericsConfig(n_periods=30.0, samples_per_period=1024, harmonic_window=(30.0, 80.0))

result = simulate(laser, target, model='rom', numerics=numerics)
print({k: v for k, v in result.diagnostics.items()})
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
plot_spectrum(result, ax=axes[0])
axes[0].set_title('ROM harmonic spectrum')
plot_pulse(result, ax=axes[1])
axes[1].set_title('Attosecond pulse (n=30..80)')
fig.tight_layout()
plt.show()
```

```python
# Fit the plateau to the BGP -8/3 scaling.
n_c = result.diagnostics['n_cutoff']
slope, intercept = result.fit_power_law(n_min=3.0, n_max=0.05 * n_c)
print(f'plateau slope = {slope:.3f}  (BGP prediction: {-8/3:.3f})')
print(f'n_cutoff      = {n_c:.1f}  (≈ γ_mirror³)')
```
