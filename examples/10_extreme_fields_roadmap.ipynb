# 10 — Extreme-fields roadmap

Reproduces the spirit of Fig. 4 of Timmis 2026: using the I_CHF/I ∝ a₀³
scaling anchored at Gemini, predict achievable CHF intensities at
flagship multi-PW facilities.  The Schwinger critical intensity is ~10²⁹
W/cm²; SEL + CHF clears it.

```python
import matplotlib.pyplot as plt
from harmonyemissions.chf.gain import predict_chf_intensity, scaling_I_chf_over_I
from harmonyemissions.units import a0_to_intensity

SCHWINGER = 2.3e29    # W/cm², critical intensity of QED
```

```python
systems = {
    'Gemini (PW)':       24.0,
    'Vulcan 20-20 (20 PW)': 70.0,
    'NSF-OPAL (20 PW)':    70.0,
    'ELI-NP (10 PW)':      60.0,
    'SEL (50 PW)':         130.0,
}
for name, a0 in systems.items():
    I_drive = a0_to_intensity(a0, 0.8)
    I_chf = predict_chf_intensity(I_drive, a0)
    print(f'{name:24s}  a₀={a0:6.1f}  I_drive={I_drive:.2e} W/cm²  I_CHF={I_chf:.2e} W/cm²  (× Schwinger: {I_chf/SCHWINGER:.2e})')
```

```python
import numpy as np
names = list(systems.keys())
a0s = np.array(list(systems.values()))
I_drives = np.array([a0_to_intensity(a, 0.8) for a in a0s])
I_chfs = np.array([predict_chf_intensity(I, a) for I, a in zip(I_drives, a0s)])

fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(a0s, I_drives, label='driver intensity', marker='o')
ax.scatter(a0s, I_chfs, label='CHF intensity', marker='*', s=120)
ax.axhline(SCHWINGER, color='red', linestyle='--', label='Schwinger limit (2.3 × 10²⁹ W/cm²)')
for name, a, I in zip(names, a0s, I_chfs):
    ax.annotate(name, (a, I), xytext=(5, 5), textcoords='offset points', fontsize=8)
ax.set_yscale('log')
ax.set_xlabel('a₀')
ax.set_ylabel('intensity [W/cm²]')
ax.set_title('Coherent Harmonic Focus pushing toward the Schwinger limit')
ax.legend(loc='upper left')
plt.show()
```
