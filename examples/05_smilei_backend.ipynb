# 05 — Dispatching to the SMILEI PIC backend

This example shows how to swap the analytical backend for SMILEI. The backend generates a 1-D input deck, dispatches `smilei`, and parses the output back into a `Result`.

**Prerequisite:** `smilei` must be on your `$PATH`. If it is not, the cell raises `SmileiNotAvailable`.

```python
import shutil
from harmonyemissions import Laser, Target, simulate
from harmonyemissions.backends.smilei import SmileiNotAvailable

laser = Laser(a0=10.0, wavelength_um=0.8, duration_fs=8.0)
target = Target.overdense(n_over_nc=200.0, gradient_L_over_lambda=0.02)

if shutil.which('smilei') is None:
    print('SMILEI not found on PATH; the next cell will raise SmileiNotAvailable.')
else:
    print('SMILEI found; dispatching.')
```

```python
try:
    result = simulate(laser, target, model='rom', backend='smilei')
    print(result.summary())
except (SmileiNotAvailable, NotImplementedError) as e:
    print(f'[expected] {type(e).__name__}: {e}')
```

The SMILEI adapter is currently wired only for the input-deck generation; the output parser is still a stub. See `docs/backends.md` for the roadmap.
