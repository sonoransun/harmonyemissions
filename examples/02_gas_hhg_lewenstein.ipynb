# 02 — Gas-phase HHG (Lewenstein / Corkum three-step)

HHG from an argon gas jet. The classical three-step model predicts a cutoff
at ħω_max = I_p + 3.17 U_p.

```python
import matplotlib.pyplot as plt
from harmonyemissions import Laser, Target, simulate
from harmonyemissions.config import NumericsConfig
from harmonyemissions.viz import plot_spectrum
```

```python
# 800-nm, ~1.4e14 W/cm² → a₀ ≈ 0.08; Argon gas.
laser = Laser(a0=0.08, wavelength_um=0.8, duration_fs=20.0, polarization='linear')
target = Target.gas('Ar')
numerics = NumericsConfig(n_periods=40.0, samples_per_period=512, harmonic_window=(17.0, 41.0))
result = simulate(laser, target, model='lewenstein', numerics=numerics)
print({k: v for k, v in result.diagnostics.items()})
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
plot_spectrum(result, ax=ax, show_fit=False)
ax.axvline(result.diagnostics['n_cutoff_corkum'], color='red', lw=0.8, ls='--', label='Corkum cutoff')
ax.set_title(f"Argon HHG, a₀={laser.a0}, λ={laser.wavelength_um} μm")
ax.legend()
plt.show()
```

```python
# Try a different gas species.
for species in ['He', 'Ne', 'Ar', 'Xe']:
    r = simulate(laser, Target.gas(species), model='lewenstein', numerics=numerics)
    print(f'{species}: I_p={r.diagnostics["ionization_potential_eV"]:.2f} eV  n_cutoff={r.diagnostics["n_cutoff_corkum"]:.1f}')
```
