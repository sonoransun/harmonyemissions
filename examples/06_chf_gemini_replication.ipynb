# 06 — Replicating the Gemini CHF experiment (Timmis et al., *Nature* 2026)

This notebook runs the `surface_pipeline` model on the paper's optimal
Gemini shot (1.2 × 10²¹ W/cm², p-pol at 45°, 2 μm super-Gaussian focal spot,
fused silica target, DPM tuned to t_HDR = 351 fs) and inspects the
harmonic spectrum, plasma dent map, and CHF gain breakdown.

```python
import matplotlib.pyplot as plt
from harmonyemissions.config import load_config
from harmonyemissions.runner import simulate_from_config
from harmonyemissions.viz import plot_beam_profile, plot_chf_gain, plot_dent_map, plot_spectrum
```

```python
cfg = load_config('../configs/chf_gemini.yaml')
result = simulate_from_config(cfg)
for k, v in result.diagnostics.items():
    print(f'{k:28s}: {v}')
print()
print('CHF gain breakdown')
for k, v in result.chf_gain.items():
    print(f'  {k:12s}: {v:.3e}')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(11, 8))
plot_spectrum(result, ax=axes[0, 0])
axes[0, 0].set_title('Spatially-averaged harmonic spectrum')
plot_beam_profile(result, which='near', ax=axes[0, 1])
plot_dent_map(result, ax=axes[1, 0])
plot_chf_gain(result, ax=axes[1, 1])
fig.tight_layout()
plt.show()
```

```python
# Fit the plateau slope to the BGP −8/3 prediction.
slope, intercept = result.fit_power_law(n_min=5, n_max=60)
print(f'plateau slope = {slope:.3f}  (BGP prediction: {-8/3:.3f})')
```
